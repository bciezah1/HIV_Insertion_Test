# Question #1

1. Produce a table ranked by the most frequently observed gene in the highest number of patient samples.


```python
# read data
import pandas as pd
df = pd.read_csv('./../../data/clinical_and_encode_overlaps.bed',sep='\t')
df.head()
```

```python
# get list of unique genes

list_of_uniq_genes =  list(set(df['gene_open'].tolist()))
list_of_uniq_genes[0:5]
```

```python
# coutn most frequent genes in highest group of patientces
list_count_times_obs_gene = []

for gene in list_of_uniq_genes:
    df_filt = df[df['gene_open'] == gene]
    count_times_obs_gene = len(list(set(df_filt['sample'].tolist())))
    list_count_times_obs_gene.append(count_times_obs_gene)
print('ready counting genes')
```

```python
# reshape data to dataframe

df_list_genes_counts = pd.DataFrame([list_of_uniq_genes,list_count_times_obs_gene])
df_list_genes_counts_t = df_list_genes_counts.transpose()
df_list_genes_counts_t = df_list_genes_counts_t.rename(columns={0: 'gene',1:'# participants'})
df_list_genes_counts_t_sorted = df_list_genes_counts_t.sort_values(by='# participants',ascending=False)
df_list_genes_counts_t_sorted = df_list_genes_counts_t_sorted[~(df_list_genes_counts_t_sorted['gene'] == '-')]
df_list_genes_counts_t_sorted = df_list_genes_counts_t_sorted.reset_index().drop('index',axis=1)
print('ready formatting table')
df_list_genes_counts_t_sorted.head(20)
```

```python
df_list_genes_counts_t_sorted.to_csv('./../../output/table_ranked_by_the_most_frequently_observed_gene_in_the_highest_number_of_patient_samples.txt',sep='\t',index=None)
print('ready saving data')
```

```python
# table summarize 

import pandas as pd

# Sample ranked data for 25 patients
ranked_df = df_list_genes_counts_t_sorted.copy()

# Define patient count thresholds and corresponding labels
thresholds = [1, 3, 5, 10, 15, 25]
labels = [ '1-3', '4-5', '6-10', '11-15', '16-25']

# Create bins and categorize genes based on patient counts
ranked_df['Patient Count Range'] = pd.cut(ranked_df['# participants'], bins=thresholds, right=True, labels=labels)

# Count the number of genes in each category
summary = ranked_df['Patient Count Range'].value_counts().reset_index()
summary.columns = ['Patient Count Range', 'Number of Genes']

# Sort the summary
summary = summary.sort_values(by='Patient Count Range')

# Display the summary
print(summary)
```
